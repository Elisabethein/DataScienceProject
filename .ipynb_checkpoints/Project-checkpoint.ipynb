```python
import numpy as np
import pandas as pd

data = pd.read_csv('heart_statlog_cleveland_hungary_final.csv')
data.head()
```

```python
unique_values = data['age'].unique()  #finds unique values in a column
unique_values
```

```python
import matplotlib.pyplot as plt


def drawPlot(column_name):
    value_counts = data[column_name].value_counts()

    plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size if needed
    plt.bar(value_counts.index, value_counts.values)
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.title('Distribution of ' + column_name)
    plt.xticks(rotation=90)  # Optional: Rotate x-axis labels if needed
    plt.show()

for column in data.columns:
    drawPlot(column)
```
